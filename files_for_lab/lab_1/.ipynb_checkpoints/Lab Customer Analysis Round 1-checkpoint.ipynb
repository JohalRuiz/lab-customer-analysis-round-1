```python
import pandas as pd
import numpy as np
```

```python
dataf1 = pd.read_csv('file1.csv')
dataf1
```

```python
dataf2 = pd.read_csv('file2.csv')
dataf2
```

```python
dataf3 = pd.read_csv('file3.csv')
dataf3
```

```python
# Determine the shape of the three dataframes:
```

```python
dataf1_shape = dataf1.shape
print("Dataframe 1 (rows, columns):", dataf1_shape)
```

```python
dataf2_shape = dataf2.shape
print("Dataframe 2 (rows, columns):", dataf2_shape)
```

```python
dataf3_shape = dataf3.shape
print("Dataframe 3 (rows, columns):", dataf3_shape)
```

```python
# Shape determination: done. Proceeding to next step
```

```python
# Header names standardization: space filling, removing capital letters, reordering columns
```

```python
#Standardize header names in lowercase and fill spaces with underscore
dataf1.columns = dataf1.columns.str.lower().str.replace(' ', '_')
dataf1.columns
```

```python
dataf1.head(3)
```

```python
#Standardize header names in lowercase and fillings spaces with underscore
dataf2.columns = dataf2.columns.str.lower().str.replace(' ', '_')
dataf2.columns
```

```python
dataf2.head(3)
```

```python
#Standardize header names in lowercase and fillings spaces with underscore
dataf3.columns = dataf3.columns.str.lower().str.replace(' ', '_')

#Rename 'state' in 'st'
dataf3 = dataf3.rename(columns={'state':'st'})
dataf3.columns
```

```python
#Reorder columns in the same order as in dataf1
dataf2 = dataf2[['customer', 'st', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
dataf2.columns
```

```python
#Reorder columns in the same order as dataf1
dataf3 = dataf3[['customer', 'st', 'gender', 'education', 'customer_lifetime_value',
       'income', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
dataf3.columns
```

```python
dataf1.columns
#Compares visually that columns are in the same order that recently modified dataf2 and dataf3 in the previous cells
```

```python
# Column rearranging: done
```

```python
dataf1.dtypes
```

```python
dataf2.dtypes
```

```python
dataf3.dtypes
```

```python
# Concatenate the three dataframes
data = pd.concat([dataf1, dataf2, dataf3])
data
```

```python
data.shape
```

```python
data.dtypes
```

```python
# Next questions:
# Which columns are numerical?
# Which columns are categorical?
```

```python
# Which columns are numerical?
data_numeric = data.select_dtypes(['int', 'float', 'int64', 'float64']).columns
print(data_numeric)
print("Number of numeric data types:", len(data_numeric))
```

```python
data_non_numeric = data.select_dtypes(exclude=['int', 'float', 'int64', 'float64']).columns
print(data_non_numeric)
print("Number of non-numeric data types:", len(data_non_numeric))
```

```python
# Which columns are categorical?
data_categoric = data.select_dtypes("object").columns
print(data_categoric)
print("Number of categoric data types:", len(data_categoric))
```

```python
#Dropping education and number of open complaints
```

```python
data1 = data.drop(['education', 'number_of_open_complaints'], axis = 1)
data1
```

```python
#checking for null values
data1[data1['customer_lifetime_value'].isna()==True]
```

```python
# Bonus cleaning: many rows in dataf1 were entirely filled with NaNs. 
# The how='all' parameter removes these rows IF they are ENTIRELY filled with NaNs.
data2 = data1.dropna(how='all')
data2
```

```python
data1_shape = data1.shape
data2_shape = data2.shape
print("Dataframe shape before removing rows entirely filled with NaN:", data1_shape)
print("Dataframe shape after removing rows entirely filled with NaN:", data2_shape)
```

```python
data2
```

```python
#checking for null values
data2[data2['customer_lifetime_value'].isna()==True]
```

```python
data2['customer_lifetime_value']
```

```python
#Also check :
# .apply(lambda x:float(str(x).replace('%','')
```

```python
#converting to string
data2['customer_lifetime_value'].to_string()
#removing %
data2['customer_lifetime_value'] = data2['customer_lifetime_value'].str.replace('%','')
#converting null fields to zero
data2['customer_lifetime_value'] = data2['customer_lifetime_value'].fillna(0)
data2
```

```python
data2.dtypes
```

```python
data3 = data2.copy()
data3['customer_lifetime_value'] =  pd.to_numeric(data3['customer_lifetime_value'], errors='coerce')
data3
```

```python
#multiplying the values in the column by 100
data3['customer_lifetime_value']=100*data3['customer_lifetime_value']
data3
```

```python
data3 = data3.drop_duplicates()
data3
```

```python
filtered_data3 = data3[(data3['customer_lifetime_value'] <= 0)]
filtered_data3
```
